```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
pd.set_option('display.max_columns', None)
```

```python
Terrorcsv = "C:\\Users\\mgeam\\Documents\\ResettlingSecurity\\TerrorismData.csv"
Terror = pd.read_csv(Terrorcsv)
Terror.head()
```

```python
Terror.dtypes
```

```python
Terror_df = Terror[['iyear', 'imonth', 'iday', 'country_txt', 'region_txt', 'provstate', 'natlty1_txt']].copy()
Terror_transform_df = Terror_df.copy()
Terror_transform_df = Terror_transform_df.rename(columns={"iyear": "Year",
                                                         "imonth": "Month",
                                                         "iday": "Day",
                                                         "country_txt": "Country",
                                                         "region_txt": "Region",
                                                         "provstate": "State",
                                                         "natlty1_txt": "Nationality "})

FinalTerror = Terror_transform_df.reset_index(drop=True)
FinalTerror
```

```python
AsyleeRefugeecsv = "C:\\Users\\mgeam\\Documents\\ResettlingSecurity\\AsyleeRefugeeData.csv"
AsyleeRefugee = pd.read_csv(AsyleeRefugeecsv)
AsyleeRefugee
```

```python
AsyleeRefugee_df = AsyleeRefugee[['Year', 'Country of origin', 'Country of asylum', 'Refugees under UNHCR’s mandate', 'Asylum-seekers']].copy()
AsyleeRefugee_transform_df = AsyleeRefugee_df.copy()
AsyleeRefugee_transform_df = AsyleeRefugee_transform_df.rename(columns={"Country of origin": "Country of Origin",
                                                         "Country of asylum": "Country of Asylum",
                                                         "Refugees under UNHCR’s mandate": "Refugees Under UNHCR’s Mandate",
                                                         "Asylum-seekers": "Asylum Seekers"})

FinalAsyleeRefugee = AsyleeRefugee_transform_df.reset_index(drop=True)
FinalAsyleeRefugee
```

```python
connection_string = "postgres:postgres@localhost:5432/Resettling Security Database"
engine = create_engine(f'postgresql://{connection_string}')
```
